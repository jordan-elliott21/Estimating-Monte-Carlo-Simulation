```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Bring in Excel Estimating Data
file_path = r'C:\Users\littl\OneDrive\Personal Data Projects\Monte_Carlo.xlsx'

df = pd.read_excel(file_path)
df = df[['Category Code', 'Category', 'COA #', 'Description', 'L $', 'ML $', 'H $', 'Calculated Value']]
df = df.drop(df.index[-1])

```

```python
# Add a column that calculates PERT

df['PERT'] = (df['L $'] + 4*df['ML $'] + df['H $']) / 6

# Ensure only 4 decimal points are shown to avoid scientific notation in results.
pd.options.display.float_format = '{:.4f}' .format

# Create sum of all rows

dfsum = df[['L $', 'ML $', 'H $']].sum()

df
```

```python
# Set the seed for the numpy random number generator to ensure random but consistent results
np.random.seed(123)

# Number of iterations for the Monte Carlo simulation
iterations = 10000

# Create an empty list to hold the results for each row
all_results = []

# Creating a function to generate a random sample from a PERT distribution, since PERT Distribution is not already included in NumPy.
def pert_random(a, m, b):
    # Calculate shape parameters for beta distribution
    alpha = 1 + 4 * ((m - a) / (b - a))
    beta = 1 + 4 * ((b - m) / (b - a))

    # Draw a random sample from a beta distribution, then scale to [a, b]
    sample = np.random.beta(alpha, beta) * (b - a) + a

    return sample

# Perform the Monte Carlo simulation for each row in the DataFrame by performing the below loop 
for index, row in df.iterrows():
    # Get the pessimistic, most likely, and optimistic estimates for this row
    pessimistic = row['L $']
    most_likely = row['ML $']
    optimistic = row['H $']

    # Create an empty list to hold the results for this row
    results = []

    # Run the simulation 'iterations' number of times by using the below imbedded loop
    for _ in range(iterations):
        # Draw a random sample from a PERT distribution defined by these estimates
        sample = pert_random(pessimistic, most_likely, optimistic)
        
        # Append the result to the list
        results.append(sample)

     # Append the list of results for this row to 'all_results'
    all_results.append(results)

# Convert 'all_results' into a DataFrame
results_df = pd.DataFrame(all_results)

results_df

```

```python
# Calculate the total cost for each iteration to find some useful statistics for the overall project
total_costs = results_df.sum(axis=0)

# Calculate some basic statistics
mean = total_costs.mean()
median = total_costs.median()
minimum = total_costs.min()
maximum = total_costs.max()
std_dev = total_costs.std()

print(f"Minimum: {minimum}")
print(f"Maximum {maximum}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")

# Generate a histogram
plt.hist(total_costs, bins=30)
plt.title('Histogram of total project costs')
plt.show()

#for i in range(0, 101, 5):
percentiles = [total_costs.quantile(i / 100) for i in range(0, 101, 5)]

#Create a DataFrame from the percentiles
percentiles_df = pd.DataFrame({'Percentile': range(0, 101, 5), 'Value': percentiles})
percentiles_df['Percentile'] = percentiles_df['Percentile'].astype(str) +'%'

# Get the value of "ML $" from dfsum
ml_dollar = dfsum.loc['ML $']

# Subtract ML $
percentiles_df['Difference'] =  percentiles_df['Value'] - ml_dollar
percentiles_df['Percentage'] = percentiles_df['Difference'] / ml_dollar

percentiles_df
```

```python
stats_df = pd.DataFrame()
table_data = pd.DataFrame()

# Calculate the stats for each original indexed row:
stats_df['Minimum'] = results_df.min(axis=1)
stats_df['Median'] = results_df.median(axis=1)
stats_df['Maximum'] = results_df.max(axis=1)
stats_df['max-med'] = stats_df['Maximum'] - stats_df['Median']
stats_df['med-min'] = stats_df['Median'] - stats_df['Minimum']

# Make a table that will be used by the Tornado Chart
table_data['Low'] = total_costs.mean() - stats_df['med-min']
table_data['Base'] = total_costs.mean()
table_data['High'] = total_costs.mean() + stats_df['max-med']
table_data['Range'] = table_data['High'] - table_data['Low']

# Join table data with original df to get the COA # and Description to use in the graph.
table_data = table_data.join(df)
table_data['COA #'] = table_data['COA #'].apply(str)
table_data['COA #_Description'] = table_data['COA #'] + " / " + table_data['Description']
table_data = table_data[['COA #_Description', 'Low', 'Base', 'High', 'Range']]
table_data
```

```python
# Sort by 'Range' in descending order and select top 10
top_10_ranges = table_data.sort_values(by='Range', ascending=False).head(10)

# Create figure and axis
fig, ax = plt.subplots()

# Plot bars (base - low on left, high - base on right)
ax.barh(y=top_10_ranges['COA #_Description'], width=top_10_ranges['Base']-top_10_ranges['Low'], left=top_10_ranges['Low'], color='red', edgecolor='black')
ax.barh(y=top_10_ranges['COA #_Description'], width=top_10_ranges['High']-top_10_ranges['Base'], left=top_10_ranges['Base'], color='blue', edgecolor='black')

# Add grid lines
ax.grid(True)

# Add a vertical line at the base value
plt.axvline(x=top_10_ranges['Base'].mean(), color='green', linestyle='--')

# Show the plot
plt.show()

top_10_ranges.sort_values(by='Range', ascending=True)
```
